Foram testados 3 algoritimos nesse relatório:

Os 3 buscam comparar sequencias de DNA e buscar a maior subsequencia parecida entre elas, a diferença principal é que o Smith Waterman é uma Heuristica, a Busca Local usa Aleatoriedade e a Busca Exaustiva gera todas as subsequencias possiveis.

### Imports:

```python
import matplotlib.pyplot as plt
import numpy as np
import time
import os
import subprocess
```

```python
def roda_com_entrada(executavel, algorithm):
    with open(algorithm) as f:
        start = time.perf_counter()
        proc = subprocess.run([executavel], input=f.read(), text=True, capture_output=True)
        end = time.perf_counter()
    return end-start

def generateTimeList(algorithm, directory):
    time_list = []
    entryList = os.listdir(directory)
    for f in entryList:
        time_list.append(roda_com_entrada(algorithm,f'{directory}/{f}'))
    return time_list, entryList

def calculateLength(directory, entryList):
    n = []
    m = []
    for file in entryList:
        with open(f'{directory}/{file}') as f :
            text_splitted = f.read().splitlines()
        n.append(len(text_splitted[0])+1)
        m.append(len(text_splitted[1])+1)
    n = [int(i) for i in n]
    m = [int(j) for j in m]
    
    return n, m

```

```python
algorithms = [
    "./Heuristica",
    "./BuscaLocal",
    "./BuscaExaustiva",
    "./BuscaExaustivaParalela"
]

directory = "./dnaSequences"

piFile = "./inPI.txt"
```

```python
durationsHeuristica, entryHeuristica = generateTimeList(algorithms[0], directory)
nHeuristica, mHeuristica = calculateLength(directory, entryHeuristica)
```

```python
durationsBuscaLocal, entryBuscaLocal = generateTimeList(algorithms[1], directory)
nBuscaLocal, mBuscaLocal = calculateLength(directory, entryBuscaLocal)
```

```python
durationsExaustiva, entryExaustiva = generateTimeList(algorithms[2], directory)
nExaustiva, mExaustiva = calculateLength(directory, entryExaustiva)
```

```python
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot_trisurf(nHeuristica, mHeuristica, durationsHeuristica, cmap="plasma")
ax1.set_title("Heuristica")
ax1.set_xlabel("DNA A")
ax1.set_ylabel("DNA B")
ax1.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax2 = fig.add_subplot(111, projection='3d')
ax2.plot_trisurf(nBuscaLocal, mBuscaLocal, durationsBuscaLocal, cmap="plasma")
ax2.set_title("Busca Local")
ax2.set_xlabel("DNA A")
ax2.set_ylabel("DNA B")
ax2.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax3 = fig.add_subplot(111, projection='3d')
ax3.plot_trisurf(nExaustiva, mExaustiva, durationsExaustiva, cmap="plasma")
ax3.set_title("Busca Exaustiva")
ax3.set_xlabel("DNA A")
ax3.set_ylabel("DNA B")
ax3.set_zlabel("Tempo")

fig.show()
```

# Resultados:

### Heuristica:
No caso da heuristica é possivel observar uma leve tendencia a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, porém ela é bem irregular então não podemos afirmar sempre a maior sequencia tem os maiores tempos, além disso podemos observar muitos picos, o que significa que para algumas sequencias em especifico ela demora muito mais.

### Busca Local:
A Busca Local é teoricamente solução mais rápida, mas por utilizar aleatoriedade o seu resultado pode se prender a máximos locais e não ser confiavel. O grafico mostra que há uma tendencia mais clara que a heuristica em direção a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, porém ainda há variações e aparentemente se aumentassemos ainda mais o numero de testes a tendencia parece que não vai para o infinito e sim para algum numero.

### Busca Exaustiva:
A Busca Exaustiva sempre retorna o resultado correto ao contrário da Local que as vezes não é o máximo global, porém é de longe a mais demorada, por checar todas as possibilidades. O grafico é clarissimo na tendencia a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, e ao contrario dos outros algoritmos não há variações nos graficos, além disso é exponencial.

## Comparação entre Exaustiva Sequencial, Paralela na CPU e paralela na GPU 

```python
with open("outGPU.txt","r",encoding = "utf-8") as file:
    lines = file.readlines()
    
mExaustivaGPU = []
nExaustivaGPU = []
durationsExaustivaGPU = []
    
for i in range(0, len(lines), 3):
    mExaustivaGPU.append(lines[i])
    nExaustivaGPU.append(lines[i+1])
    durationsExaustivaGPU.append(lines[i+2])
    
nExaustivaGPU = [int(i) for i in nExaustivaGPU]
mExaustivaGPU = [int(j) for j in mExaustivaGPU]
durationsExaustivaGPU = [float(j) for j in durationsExaustivaGPU]
```

```python
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot_trisurf(nExaustiva, mExaustiva, durationsExaustiva, cmap="plasma")
ax1.set_title("Busca Exaustiva Sequencial")
ax1.set_xlabel("DNA A")
ax1.set_ylabel("DNA B")
ax1.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax2 = fig.add_subplot(111, projection='3d')
ax2.plot_trisurf(nExaustivaP, mExaustivaP, durationsExaustivaP, cmap="plasma")
ax2.set_title("Busca Exaustiva Paralela")
ax2.set_xlabel("DNA A")
ax2.set_ylabel("DNA B")
ax2.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax3 = fig.add_subplot(111, projection='3d')
ax3.plot_trisurf(nExaustivaGPU, mExaustivaGPU, durationsExaustivaGPU, cmap="plasma")
ax3.set_title("Busca Exaustiva GPU")
ax3.set_xlabel("DNA A")
ax3.set_ylabel("DNA B")
ax3.set_zlabel("Tempo")

fig.show()
```

A busca exaustiva usando GPU continua exponencial porém podemos notar que ela é 14 vezes mais rapida aproximadamente que a sequencial para a mesma entrada, sendo um pouco mais de 2 vezes mais rapida que a paralela por CPU. Além disso 
mesmo aparentando ser exponencial a curva parece crescer em uma velocidade menor que as outras buscas.

```python
def maior(item1, item2):
    if item1 > item2:
        return item1
    else:
        return item2

sizeExaustiva = []
sizeExaustivaP = []
sizeExaustivaGPU = []
for i in range(len(nExaustiva)):
    sizeExaustiva.append(maior(nExaustiva[i], mExaustiva[i]))
for i in range(len(nExaustivaP)):    
    sizeExaustivaP.append(maior(nExaustivaP[i], mExaustivaP[i]))
for i in range(len(nExaustivaGPU)):    
    sizeExaustivaGPU.append(maior(mExaustivaGPU[i], nExaustivaGPU[i]))
    
sizeExaustiva.sort()
sizeExaustivaP.sort()
sizeExaustivaGPU.sort()
durationsExaustiva2 = durationsExaustiva
durationsExaustivaP2 = durationsExaustivaP
durationsExaustivaGPU2 = durationsExaustivaGPU
durationsExaustiva2.sort()
durationsExaustivaP2.sort()
durationsExaustivaGPU2.sort()
```

```python
fig = plt.figure(figsize=(15, 8))
plt.plot(sizeExaustiva, durationsExaustiva, label = "Sequencial", linestyle=":")
plt.plot(sizeExaustivaP, durationsExaustivaP, label = "CPU Paralela", linestyle=":")
plt.plot(sizeExaustivaGPU, durationsExaustivaGPU, label = "GPU Paralela", linestyle=":")
plt.legend()
plt.xlim(100, 150)
plt.ylim(0, 200)
plt.title("Busca Exaustiva feita de diversas maneiras")
plt.xlabel("Tamanho das sequencias")
plt.ylabel("Tempo necessario para rodar")
plt.show()
```

Primeiro podemos perceber que o grafico se encontra "cheio de pontas" isso se deve ao fato que os testes foram feitos aumentando de 5 em 5 as sequencias e se o grafico tivesse tido entradas a cada aumento individual provavelmente seria muito mais continuo.

Mas o mais interessante a se notar nesse gráfico é que para entradas pequenas a mais rápida é a CPU paralela, porém quando as entradas começam a aumentar muito a GPU paralela se torna a mais rápida. Isso provavelmente se deve ao fato que há uma demora significativa em alocar os vetores na GPU e para entradas pequenas essa demora não compensa e é mais vantajoso usar a CPU.

Por fim a CPU individual consegue sempre ser mais demorada que ambas então seu uso não valeria a pena a não ser pela simplicidade de criação de código, que porém para entradas maiores não compensa.
