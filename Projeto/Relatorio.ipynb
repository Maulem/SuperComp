Foram testados 3 algoritimos nesse relatório:
- Smith Waterman
- Busca Local
- Busca Exaustiva

Os 3 buscam comparar sequencias de DNA e buscar a maior subsequencia parecida entre elas, a diferença principal é que o Smith Waterman é uma Heuristica, a Busca Local usa Aleatoriedade e a Busca Exaustiva gera todas as subsequencias possiveis.

Os parametros para comparação utilizados foram estes:
- Match : 2
- Mismatch : -1
- Gap : -1

e usando eles é possivel obter-se uma pontuação em cada subsequencia e obter-se a pontuação máxima.

### Imports:

```python
import matplotlib.pyplot as plt
import numpy as np
import time
import os
import subprocess
```

```python
def roda_com_entrada(executavel, algorithm):
    with open(algorithm) as f:
        start = time.perf_counter()
        proc = subprocess.run([executavel], input=f.read(), text=True, capture_output=True)
        end = time.perf_counter()
    return end-start

def generateTimeList(algorithm, directory):
    time_list = []
    entryList = os.listdir(directory)
    for f in entryList:
        time_list.append(roda_com_entrada(algorithm,f'{directory}/{f}'))
    return time_list, entryList

def calculateLength(directory, entryList):
    n = []
    m = []
    for file in entryList:
        with open(f'{directory}/{file}') as f :
            text_splitted = f.read().splitlines()
        n.append(len(text_splitted[0])+1)
        m.append(len(text_splitted[1])+1)
    n = [int(i) for i in n]
    m = [int(j) for j in m]
    
    return n, m
```

```python
algorithms = [
    "./Heuristica",
    "./BuscaLocal",
    "./BuscaExaustiva"
]

directory = "./dnaSequences"

piFile = "./inPI.txt"
```

```python
durationsHeuristica, entryHeuristica = generateTimeList(algorithms[0], directory)
nHeuristica, mHeuristica = calculateLength(directory, entryHeuristica)
```

```python
durationsBuscaLocal, entryBuscaLocal = generateTimeList(algorithms[1], directory)
nBuscaLocal, mBuscaLocal = calculateLength(directory, entryBuscaLocal)
```

```python
durationsExaustiva, entryExaustiva = generateTimeList(algorithms[2], directory)
nExaustiva, mExaustiva = calculateLength(directory, entryExaustiva)
```

```python
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot_trisurf(nHeuristica, mHeuristica, durationsHeuristica, cmap="plasma")
ax1.set_title("Heuristica")
ax1.set_xlabel("DNA A")
ax1.set_ylabel("DNA B")
ax1.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax2 = fig.add_subplot(111, projection='3d')
ax2.plot_trisurf(nBuscaLocal, mBuscaLocal, durationsBuscaLocal, cmap="plasma")
ax2.set_title("Busca Local")
ax2.set_xlabel("DNA A")
ax2.set_ylabel("DNA B")
ax2.set_zlabel("Tempo")

fig = plt.figure(figsize=(15, 8))
ax3 = fig.add_subplot(111, projection='3d')
ax3.plot_trisurf(nExaustiva, mExaustiva, durationsExaustiva, cmap="plasma")
ax3.set_title("Busca Exaustiva")
ax3.set_xlabel("DNA A")
ax3.set_ylabel("DNA B")
ax3.set_zlabel("Tempo")

fig.show()
```

# Resultados:

### Heuristica:
No caso da heuristica é possivel observar uma leve tendencia a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, porém ela é bem irregular então não podemos afirmar sempre a maior sequencia tem os maiores tempos.

### Busca Local:
A Busca Local é teoricamente solução mais rápida, mas por utilizar aleatoriedade o seu resultado pode se prender a máximos locais e não ser confiavel. O grafico mostra que há uma tendencia mais clara que a heuristica em direção a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, porém ainda há variações.

### Busca Exaustiva:
A Busca Exaustiva sempre retorna o resultado correto ao contrário da Local que as vezes não é o máximo global, porém é de longe a mais demorada, por checar todas as possibilidades. O grafico é clarissimo na tendencia a demorar mais (cores mais claras) conforme o tamanho da sequencia aumenta, e ao contrario dos outros algoritmos não há variações nos graficos.
